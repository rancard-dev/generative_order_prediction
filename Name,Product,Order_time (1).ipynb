```python
import pandas as pd
import json

chunks = pd.read_json('user_orders.json', lines=True, chunksize = 100)

reviews = pd.concat([c for c in chunks])

reviews.head()
```

```python
for person in reviews['user']:
    try:
        print(person['name'])
    except:
        print(person['msisdn'])
        continue
    
```

```python
for product in reviews['products']:
    p = product[0]
    for item, name in p.items():
        if item == 'name':
            print(name, end = ', ')
        if item == 'description':
            print(name, end = ', ')
        if item == 'collection':
            c = name
            for key,value in c.items():
                if key == 'name':
                    print(value, end = ', ')
        if item == 'variants':
            var = name
            v = var[0]
            for key,value in v.items():
                if key == 'name':
                    print(value)
```

```python
def order_time(q):
    if q <=12:
        print('- Morning')
    elif q<=17:
        print('- Afternoon')
    elif q<=24:
        print('- Evening')
    


for time in reviews['_modified']:
    new_time = json.dumps(time)
    again = new_time.replace('T','  ')
    s = again[11:-11]
    q=int(s[12:14])
    print(s, end=' ')
    order_time(q)
```
